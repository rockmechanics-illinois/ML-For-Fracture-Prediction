Define current wells and where data is extracted from

```python
WELLS = ["CCS1", "VW1", "VW2"]

NORM_DIR = "REU Project/Data/Norm CSV Files/norm_"
RAW_DIR = "REU Project/Data/Raw CSV Files/raw_"
```

Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Retrieve Data

```python
norm_data = {}
raw_data = {}
for well in WELLS:
  norm_df = pd.read_csv(NORM_DIR + well + '.csv')
  raw_df = pd.read_csv(RAW_DIR + well + '.csv')
  norm_data[well] = norm_df
  raw_data[well] = raw_df
```

Get Unique Features

```python
features = list(norm_data[WELLS[0]].columns)
features.remove("FORM") #can't make box plot for non-numerical vals
```

Make Box Plot For Each Feature For Each Well (Normalized or Raw Data)

```python
def plot_feature(data_dic):
  """
  Makes a box plot for each feature across all relevant wells. This can be
  used to ensure that wells don't contain any outlier data distributions for
  any specific features.
  """
  for feat in features:
    fig, axs = plt.subplots(1, 3, figsize=(12,4))
    for i, well in enumerate(WELLS):
      axs[i].boxplot(data_dic[well][feat])
      axs[i].set_title(f"{well} : {feat}")
      axs[i].yaxis.set_visible = True
      axs[i].set_xticks([])

    plt.show()

###choose whether to plot raw or min-max normalized data
#data_dic = norm_data
data_dic = raw_data
plot_feature(data_dic)
```
